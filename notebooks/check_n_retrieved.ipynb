```python
import json

filename ="evaluate-rag-deepseek-r1-k30-chroma_2025-04-15_17-24-42.jsonl"
with open(filename, 'r') as f:
    lines = f.readlines()
    data = [json.loads(line) for line in lines]

avg_length_sources = [len(x['retrieved_documents_sources']) for x in data]
```

```python
# Calculate the average length of the sources
avg_length = sum(avg_length_sources) / len(avg_length_sources)
print(f"Average length of sources: {avg_length}")

# Make a plot
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set(style="whitegrid")
plt.figure(figsize=(6, 4))
plt.hist(avg_length_sources, bins=30, color='blue', alpha=0.7)
plt.title('Numbers of sources retrieved when k=30')
```

```python
# Loop over the data and make the same plot for each query_type
for query_type in set(x['query_type'] for x in data):
    avg_length_sources = [len(x['retrieved_documents_sources']) for x in data if x['query_type'] == query_type]
    avg_length = sum(avg_length_sources) / len(avg_length_sources)
    print(f"Average length of sources for {query_type}: {avg_length}")
    plt.hist(avg_length_sources, bins=30, color='blue', alpha=0.7)
    plt.title(f'Numbers of sources retrieved when k=30 for {query_type}')
    plt.xlabel('Number of sources')
    plt.ylabel('Frequency')
    plt.show()
```
