```python
import os
import pandas as pd
from bs4 import BeautifulSoup as bs
from collections import Counter

from t0_1.rag.evaluate import parse_deepseek_r1
```

```python
output_files = os.listdir("./evaluation_1k/")
```

```python
query_types = ["all", "cluster", "vague", "hypochondriac", "basic", "downplay"]

severities = [
    "A&E",
    "Ambulance",
    "Routine GP appointment",
    "Self-care",
    "Urgent Primary Care",
]

severities_simplified = [
    "A&E",
    "Routine GP appointment",
    "Self-care",
    "Urgent Primary Care",
]
```

```python
from ast import literal_eval


def get_predicted_severity_level(output):
    # if no tool call, return None
    if not isinstance(output, list) or len(output) == 0:
        return None

    # if multiple tool calls, return None
    if len(output) > 1:
        return None

    # else, parse the tool call
    arguments = output[0]["function"]["arguments"]
    try:
        arguments = literal_eval(arguments)
    except (SyntaxError, ValueError):
        print(f"Error evaluating arguments: {arguments}")
        return None

    return arguments.get("severity_level", None)


def get_predicted_severity_deepseek_r1(output):
    condition, severity_level = parse_deepseek_r1(output)
    return severity_level
```

```python
def get_predicted_severity_type(row):
    true_severity = row["severity_level"]
    predicted_severity = row["predicted_severity_level"]

    # if no severity level, return None
    if predicted_severity is None:
        return None

    # if severity level is not in the list, return None
    if predicted_severity not in severities:
        return None

    # else work out if the prediction is more severe or less severe than the true severity
    if true_severity == predicted_severity:
        return "correct"
    elif severities.index(predicted_severity) < severities.index(true_severity):
        return "less severe"
    else:
        return "more severe"


def get_predicted_simplified_severity_type(row):
    true_severity = row["severity_level"]
    predicted_severity = row["predicted_severity_level"]

    # convert Ambulance to A&E
    if true_severity == "Ambulance":
        true_severity = "A&E"
    if predicted_severity == "Ambulance":
        predicted_severity = "A&E"

    # if no severity level, return None
    if predicted_severity is None:
        return None

    # if severity level is not in the list, return None
    if predicted_severity not in severities:
        return None

    # else work out if the prediction is more severe or less severe than the true severity
    if true_severity == predicted_severity:
        return "correct"
    elif severities_simplified.index(predicted_severity) < severities_simplified.index(
        true_severity
    ):
        return "less severe"
    else:
        return "more severe"


severity_types = [None, "correct", "less severe", "more severe"]
```

```python
with open("../templates/rag_evaluation_system_prompt.txt") as f:
    system_prompt = f.read()
with open("../templates/rag_evaluation_system_prompt_deepseek_r1.txt") as f:
    system_prompt_deepseek_r1 = f.read()
```

```python
results = pd.DataFrame(
    columns=[
        "Embedding Method",
        "Supporting documents retrieved",
        "Query Type",
        "LLM",
        "Predicted Severity Type",
        "Count",
        "Score",
        "Count (no Ambulance)",
        "Score (no Ambulance)",
    ],
)

for output_file in output_files:
    # Get the model name and k value from the filename
    output_file_split = output_file.split("_")[0].split("-")
    model = "-".join(output_file_split[2:4]).replace("pt", ".")
    k = output_file_split[4][1:]
    print(model, k)

    # Read the jsonl file
    print(f"Reading './evaluation_1k/{output_file}'")
    data = pd.read_json(f"./evaluation_1k/{output_file}", lines=True)
    if model == "deepseek-r1":
        data["predicted_severity_level"] = data["rag_answer"].apply(
            lambda x: get_predicted_severity_deepseek_r1(x)
        )
    else:
        data["predicted_severity_level"] = data["rag_tool_calls"].apply(
            lambda x: get_predicted_severity_level(x)
        )

    data["Predicted Severity Type"] = data.apply(
        lambda x: get_predicted_severity_type(x), axis=1
    )
    data["Predicted Severity Type (no Ambulance)"] = data.apply(
        lambda x: get_predicted_simplified_severity_type(x), axis=1
    )

    # check we are sending the correct prompts i.e. with severity levels of "A&E", "Ambulance", "Routine GP appointment", "Self-care" and "Urgent Primary Care"
    for i, messages in enumerate(data["rag_message"].values):
        if not isinstance(messages, list) or len(messages) == 0:
            assert len(data.iloc[i]["error"])
            continue
        if model == "deepseek-r1":
            if messages[0] != system_prompt_deepseek_r1:
                print(f"Messages do not match for {i}: {bs(messages[0])}")
        else:
            if messages[0] != system_prompt:
                print(f"Messages do not match for {i}: {bs(messages[0])}")

    for query_type in query_types:
        if query_type == "all":
            sub_data = data
        else:
            # Get data for the specific query type
            sub_data = data[data["query_type"] == query_type]

        # Get counts for each severity type
        counts = Counter(sub_data["Predicted Severity Type"])
        # fill in empty values
        for severity_type in severity_types:
            if severity_type not in counts:
                counts[severity_type] = 0

        counts_simplified = Counter(sub_data["Predicted Severity Type (no Ambulance)"])
        # fill in empty values
        for severity_type in severity_types:
            if severity_type not in counts_simplified:
                counts_simplified[severity_type] = 0

        counts_df = pd.DataFrame.from_dict(
            dict(counts), orient="index", columns=["Count"]
        ).reset_index(names="Predicted Severity Type")
        counts_simplified_df = pd.DataFrame.from_dict(
            dict(counts_simplified), orient="index", columns=["Count (no Ambulance)"]
        ).reset_index(names="Predicted Severity Type")
        counts_df = counts_df.merge(
            counts_simplified_df, on="Predicted Severity Type", how="left"
        )
        counts_df["Supporting documents retrieved"] = k
        counts_df["Query Type"] = query_type
        counts_df["LLM"] = model
        counts_df["Score"] = counts_df["Count"] / len(sub_data)
        counts_df["Score (no Ambulance)"] = counts_df["Count (no Ambulance)"] / len(
            sub_data
        )

        results = pd.concat([results, counts_df], ignore_index=True)
```

```python
results.sort_values(
    by=[
        "Query Type",
        "Supporting documents retrieved",
        "LLM",
        "Predicted Severity Type",
    ],
    inplace=True,
)
results["Embedding Method"] = "sentence-transformers/all-mpnet-base-v2"
```

```python
results.to_csv(
    "./2b_severity_evaluation.csv",
    index=False,
)
```

```python
md_text = """
## 2b_580a395.md severity level evaluation

"""

for query_type in query_types:
    sub_results = results[results["Query Type"] == query_type]

    # add to markdown
    md_text += f"### {query_type}\n"
    for llm in sub_results["LLM"].unique():
        sub_results_llm = sub_results[sub_results["LLM"] == llm].copy(deep=True)
        sub_results_llm.fillna("NA", inplace=True)

        # add to markdown
        md_text += f"#### {llm}\n"
        md_text += sub_results_llm.to_markdown(
            index=False, tablefmt="github", floatfmt=".2f"
        )
        md_text += "\n\n"

    md_text += "\n\n"

with open("./2b_severity_evaluation_580a395.md", "w") as f:
    f.write(md_text)
```
